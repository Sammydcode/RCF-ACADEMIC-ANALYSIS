**TOP 10 STUDENT WITH HIGH PERFORMANCES (IDENTIFIED BY DEPARTMENT)**

```python
df
```

```python
top_10_high_performance = (df.sort_values(by = "CURRENT CGPA", ascending = False).head(10)).loc[:, ["NAME (SURNAME FIRST)","DEPARTMENT IN SCHOOL", "CURRENT CGPA", "LEVEL"]]
top_10_high_performance
#Identifying the top 10 student with their name, department in school and their cgpa and level
```

**INSIGHT**

From the above analysis, it is observed that a "100 LEVEL" student is having a cgpa which is not supposed to be so until "200 LEVEL". Hence any "100 LEVEL" with a cgpa greater than 0 would be dropped.

**DROPPING RECORDS OF STUDENT IN 100 LEVEL WITH A CGPA GRATER THAN 0**

```python
df
```

```python
df.drop(df[(df["LEVEL"] == 100) & (df["CURRENT CGPA"] > 0)].index, inplace = True)
df
#dropping the records of 100 level student with a cgpa greater than 0
```

**IDENTIFYING THE TOP 5 STUDENT WITH THE HIGHEST PERFORMANCE AGAIN AFTER ALTERING THE DATASET**

```python
top_10_high_performance = (df.sort_values(by = "CURRENT CGPA", ascending = False).head(10)).loc[:, ["NAME (SURNAME FIRST)","DEPARTMENT IN SCHOOL", "CURRENT CGPA", "LEVEL"]]
top_10_high_performance
#Identifying the top 5 student with their name, department in school and their cgpa and level
```

```python
import matplotlib.pyplot as plt
```

```python
#visualizing with a line plot
plt.figure(figsize = (10, 6))
plt.plot(top_10_high_performance["NAME (SURNAME FIRST)"],top_10_high_performance["CURRENT CGPA"], color = "red", marker = "o")
plt.xticks(rotation = 90)
plt.title("TOP 10 STUDENT WITH HIGH PERFORMANCES", fontweight = "bold", fontsize = 14)
plt.ylabel("CGPA", fontweight = "bold")
plt.xlabel("NAME OF STUDENTS", fontweight = "bold")
plt.grid()
plt.show()
```

**TOP 10 STUDENT WITH LOW PERFORMANCES THAT IS NOT EQUALS TO ZERO (IDENTIFIED BY DEPARTMENT)**

```python
top_10_low_performance = ((df.sort_values(by = "CURRENT CGPA", ascending = False).loc[df["CURRENT CGPA"] != 0]).tail(10)).loc[:, ["NAME (SURNAME FIRST)","DEPARTMENT IN SCHOOL", "CURRENT CGPA", "LEVEL"]]
top_10_low_performance

#Identifying the top 10 lowest student with their name, department in school and their cgpa and level
```

```python
#visualizing with a line plot
plt.figure(figsize = (10, 6))
plt.plot(top_10_low_performance["NAME (SURNAME FIRST)"],top_10_low_performance["CURRENT CGPA"], color = "red", marker = "o")
plt.xticks(rotation = 90)
plt.title("TOP 10 STUDENT WITH LOW PERFORMANCES", fontweight = "bold", fontsize = 14)
plt.ylabel("CGPA", fontweight = "bold")
plt.xlabel("NAME OF STUDENTS", fontweight = "bold")
plt.grid()
plt.show()
```

**IDENTIFYING MOST POPULAR ISSUES FACED BY THE STUDENTS**

```python
df
```

```python
df.columns
```

```python
popular_issues = df.groupby("SUMMARY ISSUES")["SUMMARY ISSUES"].count()
popular_issues
```

```python
#Visualizing the statistics
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (10, 6))
plt.bar(popular_issues.index,popular_issues.values, color = "blue")
plt.title("STUDENT ISSUES STATISTICS", fontweight = "bold", fontsize = 14)
plt.xlabel("STUDENT ISSUES", fontsize = 12)
plt.ylabel("FREQUENCY", fontsize = 12)
plt.show()
```

**INSIGHT**

From the above analysis, it is observed that the most popular issues among the student is reading issues followed by balancing issues and then time management.

The "other issues" statistics describes issues that are genuinely due to school factors like issues from lecturers and so on.

It is also observed that majority of the student were not having any issue or challenge.

**DEPARTMENT WITH THE TOP PERFORMANCE**

```python
df
```

```python
#Grouping each department by their average cgpa
dept_by_cgpa = df.groupby("DEPARTMENT IN SCHOOL")["CURRENT CGPA"].mean().sort_values(ascending = False)
dept_by_cgpa
```

```python
#Bringing out the top 10 department by performance
dept_by_cgpa_10 = pd.DataFrame(dept_by_cgpa.head(10))
dept_by_cgpa_10
```

```python
#Visulizing with a line chart
plt.figure(figsize = (10, 6))
plt.plot(dept_by_cgpa_10, color = "red", marker = "o")
plt.xticks(rotation = 90)
plt.title("TOP DEPARTMENT BY THEIR AVERAGE CGPA", fontweight = "bold", fontsize = 14)
plt.ylabel("CGPA", fontweight = "bold")
plt.xlabel("Departments", fontweight = "bold")
plt.grid()
plt.show()
```

**DEPARTMENT WITH THE LEAST PERFORMANCE**

```python
dept_by_cgpa_low10 = pd.DataFrame(dept_by_cgpa.tail(10))
dept_by_cgpa_low10 
```

```python
#Visulizing with a line chart
plt.figure(figsize = (10, 6))
plt.plot(dept_by_cgpa_low10, color = "red", marker = "o")
plt.xticks(rotation = 90)
plt.title("LEAST DEPARTMENT BY THEIR AVERAGE CGPA", fontweight = "bold", fontsize = 14)
plt.ylabel("CGPA", fontweight = "bold")
plt.xlabel("Departments", fontweight = "bold")
plt.grid()
plt.show()
```

**DEPARTMENT WITH THE LEAST PERFORMANCE WHERE THEIR CGPA IS NOT EQUALS TO ZERO**

```python
#Bringing out the top lowest 10 department by performance excluding the one with zero cgpa
dept_by_cgpa_low10_2 = pd.DataFrame((dept_by_cgpa.loc[dept_by_cgpa != 0]).tail(10))
dept_by_cgpa_low10_2
```

```python
#Visulizing with a line chart
plt.figure(figsize = (10, 6))
plt.plot(dept_by_cgpa_low10_2, color = "red", marker = "o")
plt.xticks(rotation = 90)
plt.title("LEAST DEPARTMENT BY THEIR AVERAGE CGPA WITH CGPA > 0", fontweight = "bold", fontsize = 14)
plt.ylabel("CGPA", fontweight = "bold")
plt.xlabel("Departments", fontweight = "bold")
plt.grid()
plt.show()
```
